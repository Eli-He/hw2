```python
inp = "x 2 , x 1 , x 3 , x 8 , x 13 , x 6 , x 5 , x 7 , x 9 , x 12 , x 10 , x 11 , x 15 , x 4 , x 14 , x 16 , x 17 , x 20 , x 18 , x 19"

inp=[int(i) for i in inp.split() if i.isdigit()]
```

```python
x=[0]
y=[0]


for i in inp:
    if i<=12:
        y.append(y[-1]+1)
        x.append(x[-1])
    else:
        y.append(y[-1])
        x.append(x[-1]+1)



plt.gca().set_aspect('equal', adjustable='box')
#plt.axis('equal')
plt.grid()
plt.ylabel('TP')
plt.xlabel('FP')
plt.scatter(x,y)

plt.xlim(x[0]-.1,x[-1]+.1)
plt.ylim(x[0]-.1,y[-1]+.1)
plt.savefig('1.pdf')
plt.show()
```

```python
w=np.array([2,1,3])
t=12
inp=np.array(
[[2, 2, 3 ],[3, 3, 2 ],[1, 2, 3 ],[1, 4, 1 ],
[4, 4, 4 ],[2, 2, 2 ],[1, 1, 1 ],[0, 4, 2 ],
[4, 0, 0 ],[3, 3, 1 ],[3, 3, 3 ]])
label=np.array([1 if i <6 else -1 for i in range(len(inp))])
margin=label*(inp.dot(w.T)-t)/np.linalg.norm(w)
result=[]
result.append(margin)
result.append((margin<0).astype(int))#0-1
result.append(np.maximum(1-margin,0))#hinge
result.append((1-margin)**2)#square
pd.DataFrame(result,index=['margin','0-1','hinge','square']).T
```

```python
inp=pd.Series([3,5,5,2,3,1,5,3,5,5,2,3,5,5,1,2,3,3,5,3,3,5,5,2,5])
result=[]
relative=inp.value_counts(normalize=True,sort=False).reindex(range(1,6),fill_value=0)
result.append(relative)
nominator=(inp.value_counts().reindex(range(1,6),fill_value=0)+1)
laplace=nominator/sum(nominator)
result.append(laplace)
result.append(laplace)
nominator=(inp.value_counts().reindex(range(1,6),fill_value=0)+4)
m_est=nominator/sum(nominator)
result.append(m_est)
result=pd.DataFrame(result,index=['relative','laplace','1-est','5-est']).T
for i in result:
    result[i].plot()
    plt.xticks(range(1,6))
    plt.title(i)
    plt.grid(True)
    plt.savefig(i+'.eps')
    plt.show()
```
